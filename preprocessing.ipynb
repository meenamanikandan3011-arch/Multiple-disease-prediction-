```python
import pandas as pd 
```

```python
# -----------------------------
# LOAD DATASETS
# -----------------------------
parkinsons_df = pd.read_csv("parkinsons - parkinsons.csv")
kidney_df = pd.read_csv("kidney_disease - kidney_disease.csv")
liver_df = pd.read_csv("indian_liver_patient - indian_liver_patient.csv")
```

```python
# -----------------------------
# DISPLAY BASIC INFO
# -----------------------------
print("Parkinsons Dataset Shape:", parkinsons_df.shape)
print("Kidney Disease Dataset Shape:", kidney_df.shape)
print("Liver Disease Dataset Shape:", liver_df.shape)
```

```python
# -----------------------------
# VIEW SAMPLE DATA
# -----------------------------
print("\nParkinsons Dataset Sample:")
print(parkinsons_df.head())

print("\nKidney Disease Dataset Sample:")
print(kidney_df.head())

print("\nLiver Disease Dataset Sample:")
print(liver_df.head())
```

ðŸ”¹ A. Handling Missing Data (Imputation)

```python
import numpy as np

# Replace '?' with NaN
kidney_df.replace("?", np.nan, inplace=True)

# Convert numeric columns properly
for col in kidney_df.columns:
    kidney_df[col] = pd.to_numeric(kidney_df[col], errors='ignore')

# Fill numerical missing values with mean
kidney_df.fillna(kidney_df.mean(numeric_only=True), inplace=True)

# Fill categorical missing values with mode
for col in kidney_df.select_dtypes(include='object'):
    kidney_df[col].fillna(kidney_df[col].mode()[0], inplace=True)
kidney_df
```

```python
# Fill missing numerical values with mean
liver_df.fillna(liver_df.mean(numeric_only=True), inplace=True)
liver_df
```

```python
parkinsons_df.isnull().sum()
```

ðŸ”¹ B. Encoding Categorical Variables (Using Pandas)

```python
from sklearn.preprocessing import LabelEncoder 
# Initialize LabelEncoder
le = LabelEncoder()

# Encode all categorical columns
for col in kidney_df.select_dtypes(include='object'):
    kidney_df[col] = le.fit_transform(kidney_df[col])
kidney_df
```

```python
le = LabelEncoder()

liver_df['Gender'] = le.fit_transform(liver_df['Gender'])
parkinsons_df['name'] = le.fit_transform(parkinsons_df['name'])
parkinsons_df
```
